```python
import PyPDF2
from google.colab import files
import os

pdf_path = "Business_Proposal.pdf"

if not os.path.exists(pdf_path):
    print(f"File '{pdf_path}' not found. Please upload the file 'Business_Proposal.pdf'.")
    try:
        uploaded = files.upload()
        if not uploaded:
            print("No file was uploaded.")
        elif pdf_path not in uploaded:
            print(f"Warning: The uploaded file(s) are not named 'Business_Proposal.pdf'. Found: {list(uploaded.keys())}")
            print("Please ensure you upload 'Business_Proposal.pdf' or update 'pdf_path' variable if you uploaded a different file.")
    except Exception as e:
        print(f"An error occurred during file upload: {e}")

# Proceed with opening the file. If it still doesn't exist, FileNotFoundError will be raised by the open() call.
with open(pdf_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    all_text = "\n".join(
        page.extract_text()
        for page in reader.pages
        if page.extract_text()
    )

with open("business_proposal_all.txt", "w", encoding="utf-8") as out:
    out.write(all_text)

print("Done extracting all pages")
```
