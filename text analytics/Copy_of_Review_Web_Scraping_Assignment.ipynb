```python
"""
Name: Alhathal Naif Mashaan D
Student ID: SW01084553
Assignment: Amazon Review Scraper
Product: Amazon Fire TV Stick 4K Max (ASIN B08MQZXN1X)
"""

# Import libraries for web requests, HTML parsing, and data storage
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Base URL for review pages of the chosen product
base_url = "https://www.amazon.com/product-reviews/B08MQZXN1X?pageNumber="

# Lists to store extracted review data
reviewer_names = []
review_dates = []
review_texts = []

# Loop through 5 pages of reviews
for page in range(1, 6):
    print(f"Scraping page {page}...")

    # Build full URL for the current review page
    url = base_url + str(page)

    # Headers to mimic a web browser (reduces chance of blocking)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36"
    }

    # Send GET request
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find each review container on the page
    reviews = soup.find_all("div", {"data-hook": "review"})

    # Extract data from each review
    for review in reviews:
        # Get reviewer name
        name_tag = review.find("span", class_="a-profile-name")
        reviewer_names.append(name_tag.text.strip() if name_tag else "N/A")

        # Get review date
        date_tag = review.find("span", {"data-hook": "review-date"})
        review_dates.append(date_tag.text.strip() if date_tag else "N/A")

        # Get review text
        body_tag = review.find("span", {"data-hook": "review-body"})
        review_texts.append(body_tag.text.strip() if body_tag else "N/A")

    # Pause to avoid being blocked
    time.sleep(2)

# Build a DataFrame from the collected data
df = pd.DataFrame({
    "Reviewer Name": reviewer_names,
    "Review Date": review_dates,
    "Review Content": review_texts
})

# Save the DataFrame to CSV
csv_filename = "amazon_fire_tv_stick_reviews.csv"
df.to_csv(csv_filename, index=False)

print(f"Scraping finished. Data saved to {csv_filename}")
```
